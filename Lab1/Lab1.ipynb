### Початкове завантаження бібліотек та набору даних

```python
# Завантаження датасету з kaggle для виконання лабораторної роботи
import kagglehub

# Download latest version
path = kagglehub.dataset_download("brendan45774/test-file")

print("Path to dataset files:", path)
```

```python
# Додавання потрібних біблотек для роботи
import pandas as pd
```

```python
# Перевірка датасету
titanic_data = pd.read_csv("titanic.csv")
titanic_data.head(10)
```

```python
# Використання методу describe()
titanic_data.describe()
```

```python
# Використання методу mean() за відповідним стовпцем до числового ряду
titanic_data["Age"].mean()
```

```python
# Використання методу unique() до числового ряду
titanic_data["Age"].unique()
```

```python
# Використання методу unique() до рядкового ряду
titanic_data["Name"].unique()
```

```python
# Використання методу value_counts() до числового ряду
titanic_survived = titanic_data.Survived.value_counts()
titanic_survived
```

```python
# Використання методу value_counts() до числового ряду у відсотках кількості виживших
titanic_survived = titanic_data.Survived.value_counts(normalize=True).reset_index(name='Percentage')
titanic_survived
```

```python
# Використання методу map() до числового ряду Fare та піднемення кожного значення у квадрат
titanic_data["Fare"].map(lambda x: x**2)
```

```python
# Використання методу apply()
titanic_data[["Age", "Fare"]].apply(lambda col: col.max())
```
