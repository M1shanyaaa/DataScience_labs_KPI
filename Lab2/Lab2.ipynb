## Мета лабораторної роботи: 
#### Навчитися швидко та ефективно вибирати необхідні дані з набору даних з використанням структур даних та інструментів бібліотеки Pandas. 


## Практична реалізація: 
#### Набути практичних навичок вибору даних з структур Sources та DataFrame.  

```python
# Додавання потрібних біблотек для роботи
import pandas as pd
```

```python
# Завантаження та перевірка датасету
titanic_data = pd.read_csv('C:\\Users\\Mishanya\\PycharmProjects\\Foundation_KPI_DataScience\\Lab1\\titanic.csv')
titanic_data
```

## 1. Вбудовані способи доступу

```python
# A) Доступ до стовпця можна забезпечити шляхом застосування його назви як атрибуту об’єкта з точковою нотацією.
titanic_data.Name
```

```python
# B) Доступ до стовпця як до елемента словника. У цьому випадку ключом словника є назва стовпця, а значенням – Series стовпця.
titanic_data["Name"]
```

```python
#C)Доступ до елемента вибраного стовпця по індексу
titanic_data["Name"][3]
```

## 2. Індексно‐орієнтований вибір у бібліотеці Pandas

```python
#А) Вибір рядка на основі індексу
titanic_data.iloc[3]
```

```python
# Б) Вибір стовпця на основі індексу 
titanic_data.iloc[:, 3]
```

```python
#В) Вибір стовпця на основі індексу з одночасною операцією зрізу з елементів стовпця в Series. 
titanic_data.iloc[:4, 3]
```

```python
#Г) Вибір зрізу рядків  
titanic_data.iloc[2:7]
```

```python
# Д) Вибір зрізу рядків  з кінця набору даних з використанням від’ємного індексу 
titanic_data.iloc[-3:]
```

## 3. Мітко‐орієнтований вибір в бібліотеці Pandas 
##### Для виконання мітко-орієнтованого вибору застосовують метод loc. 

```python
#А) Вибір по індексу елементу стовпця заданого по імені. 
titanic_data.loc[3, 'Name']
```

```python
# Б) Вибір підмножини стовпців, імена яких задані списком 
titanic_data.loc[:,['Name', 'Survived','Age']]
```

```python
#В) Вибір перших 7 рядків з підмножини стовпців, імена яких задані списком
titanic_data.loc[:6,['Name', 'Survived','Age']]
```

```python
#Г) Вибір рядків з кроком у заданому діапазоні з підмножини стовпців, імена яких задані списком
titanic_data.loc[1:18:2,['Name', 'Survived','Age']]
```

## 4. Маніпулювання індексом 

```python
titanic_data.set_index("Name")
```

## 5. Умовний вибір 

```python
#A) Проста умова 
age_mean = titanic_data.Age.mean()
print("Середній вік на борту: ", age_mean)
titanic_data.loc[titanic_data.Age >= age_mean]
```

```python
# Б) Складна умова (вижевші у трагедії, вік яких вищий за середній)
age_mean = titanic_data.Age.mean()
print("Середній вік на борту: ", age_mean)
titanic_data.loc[(titanic_data.Age > age_mean) & (titanic_data.Survived == 1)]
```

```python
# В) Умовні селектори 
titanic_data.loc[(titanic_data.Sex == "female") & (titanic_data.Age.isin([50,51,52,53]))]   # жінки, яким 50-53 років на момент катастрофи
```

```python
# дозволяють виділити значення, які є (або не є) порожніми (NaN). 
titanic_data.loc[titanic_data.Ticket.isnull()] 
```

## 6. Редагування даних 

```python
# Присвоєння значення елементам DataFrame.
titanic_data["Birth"] = titanic_data["Age"] + 1
titanic_data
```
